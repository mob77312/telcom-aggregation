## 3. Aggregation Function

```python
import pandas as pd
rakamin_class = pd.read_csv('rakamin_class.csv')
```

```python
# group by province, tiap kolom dicari 2 statistics

rakamin_class.groupby(['province']).agg( [ 'count' , 'mean']).reset_index()
```

```python
# group by class, satu kolom dicari 3 statistics berbeda

rakamin_class.groupby(['province']).agg( {'last_exam_score' : ['max'],
                                          'age' : ['mean'] } ).reset_index()
```

```python
# group by class, satu kolom dicari 3 statistics berbeda

rakamin_group = rakamin_class.groupby(['class']).agg( {'age' : ['max','min','mean'] } )
rakamin_group
```

```python
rakamin_group = rakamin_group.reset_index()

rakamin_group.columns = ['class','age_max','age_min','age_rata_rata']
```

```python
rakamin_group
```

```python
# simpan hasil group by

class_group = rakamin_class.groupby(['class']).agg( {'age' : ['max','min','mean'] }).reset_index()
class_group
```

```python
# UBAH nama kolom terlebih dahulu

class_group.columns = ['class','age_max','age_min','age_mean'] # SESSION 13 - UBAH SEMUA NAMA KOLOM
class_group
```

```python
class_group[class_group['age_max']==29]
```

```python
# CONTOH LAIN

class_group = rakamin_class.groupby(['class']).agg( {'age' : ['max','min','mean'],
                                                     'last_exam_score' : ['max','min'],
                                                     'active' : ['max','min']}).reset_index()

class_group
```

```python
class_group.columns = ['class','age_max','age_min','age_mean','score_max','score_min','active_max','active_mean']
class_group
```

2 Output di atas sama saja

```python
telecom_grouped = telecom.groupby(['Occupation','Churn']).agg({'CustomerID' : ['count']}).reset_index()
telecom_grouped.columns = ['Occupation','Churn','customers']
telecom_grouped
```

```python
# membuat kolom baru total_customers, yang merupakan penjumlahan customers per occupation

telecom_grouped['total_customers'] = telecom_grouped.groupby(['Occupation'])['customers'].transform('sum')
telecom_grouped
```

```python
# membuat kolom percentage yang merupakan customers/total_customers

telecom_grouped['percentage'] = telecom_grouped['customers']/telecom_grouped['total_customers']
telecom_grouped
```

```python
# filte ryang churn = yes saja

telecom_grouped[telecom_grouped['Churn']=='Yes']

telecom_grouped
```
